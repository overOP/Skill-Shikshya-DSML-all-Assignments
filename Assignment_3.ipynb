```python
from functools import reduce

students = ["Ram", "Sita", "Hari", "Gita", "Shyam", "Rita"]
marks = [85, 42, 78, 91, 35, 67]

print("Student List:")
for sn, name in enumerate(students, start=1):
    print(sn, "-", name)

records = list(zip(students, marks))

print("\nStudent Records:")
for record in records:
    print(record)

def get_grade(mark):
    if mark >= 90:
        return "A+"
    elif mark >= 80:
        return "A"
    elif mark >= 70:
        return "B+"
    elif mark >= 60:
        return "B"
    elif mark >= 50:
        return "C+"
    elif mark >= 40:
        return "C"
    else:
        return "NG"

grades = list(map(get_grade, marks))

print("\nGrades:")
for name, grade in zip(students, grades):
    print(name, ":", grade)

passed_students = list(filter(lambda record: record[1] > 40, records))

print("\nPassed Students:")
for student in passed_students:
    print(student)

total_marks = reduce(lambda x, y: x + y, marks)

print("\nTotal Marks =", total_marks)

highest_mark = reduce(lambda x, y: x if x > y else y, marks)

print("Highest Mark =", highest_mark)
```
