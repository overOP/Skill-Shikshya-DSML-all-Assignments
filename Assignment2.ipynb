```python
#Student Result Management System
def student_result():
    students = {}

    name1 = input("Enter name of student 1: ")
    marks1 = float(input("Enter marks of " + name1 + ": "))

    name2 = input("Enter name of student 2: ")
    marks2 = float(input("Enter marks of " + name2 + ": "))

    name3 = input("Enter name of student 3: ")
    marks3 = float(input("Enter marks of " + name3 + ": "))

    name4 = input("Enter name of student 4: ")
    marks4 = float(input("Enter marks of " + name4 + ": "))

    name5 = input("Enter name of student 5: ")
    marks5 = float(input("Enter marks of " + name5 + ": "))

    students[name1] = marks1
    students[name2] = marks2
    students[name3] = marks3
    students[name4] = marks4
    students[name5] = marks5

    # Student 1
    if marks1 >= 80:
        grade1 = "A"
    elif marks1 >= 60:
        grade1 = "B"
    elif marks1 >= 40:
        grade1 = "C"
    else:
        grade1 = "F"
    print("Name:", name1, "| Marks:", marks1, "| Grade:", grade1)

    # Student 2
    if marks2 >= 80:
        grade2 = "A"
    elif marks2 >= 60:
        grade2 = "B"
    elif marks2 >= 40:
        grade2 = "C"
    else:
        grade2 = "F"
    print("Name:", name2, "| Marks:", marks2, "| Grade:", grade2)

    # Student 3
    if marks3 >= 80:
        grade3 = "A"
    elif marks3 >= 60:
        grade3 = "B"
    elif marks3 >= 40:
        grade3 = "C"
    else:
        grade3 = "F"
    print("Name:", name3, "| Marks:", marks3, "| Grade:", grade3)

    # Student 4
    if marks4 >= 80:
        grade4 = "A"
    elif marks4 >= 60:
        grade4 = "B"
    elif marks4 >= 40:
        grade4 = "C"
    else:
        grade4 = "F"
    print("Name:", name4, "| Marks:", marks4, "| Grade:", grade4)

    # Student 5
    if marks5 >= 80:
        grade5 = "A"
    elif marks5 >= 60:
        grade5 = "B"
    elif marks5 >= 40:
        grade5 = "C"
    else:
        grade5 = "F"
    print("Name:", name5, "| Marks:", marks5, "| Grade:", grade5)

    highest = marks1
    if marks2 > highest:
        highest = marks2
    if marks3 > highest:
        highest = marks3
    if marks4 > highest:
        highest = marks4
    if marks5 > highest:
        highest = marks5

    lowest = marks1
    if marks2 < lowest:
        lowest = marks2
    if marks3 < lowest:
        lowest = marks3
    if marks4 < lowest:
        lowest = marks4
    if marks5 < lowest:
        lowest = marks5

    total = marks1 + marks2 + marks3 + marks4 + marks5
    average = total / 5

    print("Highest Marks:", highest)
    print("Lowest Marks :", lowest)
    print("Average Marks:", average)

student_result()
```

```python
#Student Result Management System
def shopping_cart():
    prices = []

    price1 = float(input("Enter price of product 1: "))
    price2 = float(input("Enter price of product 2: "))
    price3 = float(input("Enter price of product 3: "))
    price4 = float(input("Enter price of product 4: "))
    price5 = float(input("Enter price of product 5: "))

    prices.append(price1)
    prices.append(price2)
    prices.append(price3)
    prices.append(price4)
    prices.append(price5)

    print("Product 1: Rs.", price1)
    print("Product 2: Rs.", price2)
    print("Product 3: Rs.", price3)
    print("Product 4: Rs.", price4)
    print("Product 5: Rs.", price5)

    total = 0
    for price in prices:
        total = total + price

    print("Total Amount: Rs.", total)

    if total >= 5000:
        discount_percent = 15
    elif total >= 3000:
        discount_percent = 10
    elif total >= 1000:
        discount_percent = 5
    else:
        discount_percent = 0

    discount_amount = (discount_percent / 100) * total
    final_amount = total - discount_amount

    print("Total Amount    : Rs.", total)

    if discount_percent == 0:
        print("Discount        : No Discount")
    else:
        print("Discount        :", discount_percent, "% off")

    print("Discount Amount : Rs.", discount_amount)
    print("Final Amount    : Rs.", final_amount)

shopping_cart()
```
