```python
from abc import ABC, abstractmethod
from datetime import datetime

# ABSTRACT BASE CLASS - Student

class Student(ABC):
    """
    Abstract base class representing a Student.
    Demonstrates: Abstraction, Encapsulation
    """

    def __init__(self, student_id, name):
        """
        Initialize student with ID and Name.
        Private attribute __marks for encapsulation.
        """
        self.__student_id = student_id
        self.__name = name
        self.__marks = []

    # --- Getters and Setters (Encapsulation) ---

    def get_student_id(self):
        """Get student ID"""
        return self.__student_id

    def get_name(self):
        """Get student name"""
        return self.__name

    def set_marks(self, subject, marks):
        """
        Set marks for a subject.
        Validates that marks are between 0-100.
        """
        if not (0 <= marks <= 100):
            print(f"❌ Invalid marks! Marks must be between 0-100")
            return False

        # Check if subject already exists and update
        for i, (subj, _) in enumerate(self.__marks):
            if subj == subject:
                self.__marks[i] = (subject, marks)
                print(f"✓ Marks updated for {subject}: {marks}")
                return True

        # Add new subject marks
        self.__marks.append((subject, marks))
        print(f"✓ Marks added for {subject}: {marks}")
        return True

    def get_marks(self):
        """Get all marks as a dictionary"""
        return {subject: marks for subject, marks in self.__marks}

    def get_average_marks(self):
        """Calculate and return average marks"""
        if not self.__marks:
            return 0
        total = sum(marks for _, marks in self.__marks)
        return round(total / len(self.__marks), 2)

    # ---- Abstract Methods (Abstraction) ----

    @abstractmethod
    def calculate_grade(self):
        """
        Abstract method to calculate grade.
        Must be implemented by subclasses.
        """
        pass

    @abstractmethod
    def display_details(self):
        """
        Abstract method to display student details.
        Must be implemented by subclasses.
        """
        pass

# CONCRETE CLASSES - Different Student Categories (Inheritance)

class UndergraduateStudent(Student):
    """
    Undergraduate student class.
    Demonstrates: Inheritance, Polymorphism
    """

    def __init__(self, student_id, name, department):
        super().__init__(student_id, name)
        self.__department = department
        self.__category = "Undergraduate"

    def calculate_grade(self):
        """Calculate grade for undergraduate (A, B, C, D, F)"""
        avg = self.get_average_marks()

        if avg >= 90:
            return "A"
        elif avg >= 80:
            return "B"
        elif avg >= 70:
            return "C"
        elif avg >= 60:
            return "D"
        else:
            return "F"

    def display_details(self):
        """Display undergraduate student details"""
        print("\n" + "="*60)
        print(f"{'UNDERGRADUATE STUDENT DETAILS':^60}")
        print("="*60)
        print(f"Student ID       : {self.get_student_id()}")
        print(f"Name             : {self.get_name()}")
        print(f"Department       : {self.__department}")
        print(f"Category         : {self.__category}")
        print(f"\nSubject Marks    : {self.get_marks()}")
        print(f"Average Marks    : {self.get_average_marks()}")
        print(f"Grade            : {self.calculate_grade()}")
        print("="*60 + "\n")


class PostgraduateStudent(Student):
    """
    Postgraduate student class.
    Demonstrates: Inheritance, Polymorphism
    """

    def __init__(self, student_id, name, specialization, thesis_topic):
        super().__init__(student_id, name)
        self.__specialization = specialization
        self.__thesis_topic = thesis_topic
        self.__category = "Postgraduate"

    def calculate_grade(self):
        """Calculate grade for postgraduate (Distinction, Merit, Pass, Fail)"""
        avg = self.get_average_marks()

        if avg >= 85:
            return "Distinction"
        elif avg >= 75:
            return "Merit"
        elif avg >= 65:
            return "Pass"
        else:
            return "Fail"

    def display_details(self):
        """Display postgraduate student details"""
        print("\n" + "="*60)
        print(f"{'POSTGRADUATE STUDENT DETAILS':^60}")
        print("="*60)
        print(f"Student ID       : {self.get_student_id()}")
        print(f"Name             : {self.get_name()}")
        print(f"Specialization   : {self.__specialization}")
        print(f"Thesis Topic     : {self.__thesis_topic}")
        print(f"Category         : {self.__category}")
        print(f"\nSubject Marks    : {self.get_marks()}")
        print(f"Average Marks    : {self.get_average_marks()}")
        print(f"Grade            : {self.calculate_grade()}")
        print("="*60 + "\n")


class InternationalStudent(Student):
    """
    International student class.
    Demonstrates: Inheritance, Polymorphism
    """

    def __init__(self, student_id, name, country, visa_status):
        super().__init__(student_id, name)
        self.__country = country
        self.__visa_status = visa_status
        self.__category = "International"

    def calculate_grade(self):
        """Calculate grade for international students (1-10 scale)"""
        avg = self.get_average_marks()
        return round((avg / 10), 1)

    def display_details(self):
        """Display international student details"""
        print("\n" + "="*60)
        print(f"{'INTERNATIONAL STUDENT DETAILS':^60}")
        print("="*60)
        print(f"Student ID       : {self.get_student_id()}")
        print(f"Name             : {self.get_name()}")
        print(f"Country          : {self.__country}")
        print(f"Visa Status      : {self.__visa_status}")
        print(f"Category         : {self.__category}")
        print(f"\nSubject Marks    : {self.get_marks()}")
        print(f"Average Marks    : {self.get_average_marks()}")
        print(f"Grade (Scale 10) : {self.calculate_grade()}")
        print("="*60 + "\n")

# STUDENT MANAGEMENT SYSTEM

class StudentManagementSystem:
    """
    System to manage students and perform operations.
    Demonstrates: Composition, Encapsulation
    """

    def __init__(self):
        self.__students = []

    def add_student(self, student):
        """Add a student to the system"""
        self.__students.append(student)
        print(f"✓ Student {student.get_name()} added successfully!")

    def find_student(self, student_id):
        """Find a student by ID"""
        for student in self.__students:
            if student.get_student_id() == student_id:
                return student
        return None

    def display_all_students(self):
        """Display details of all students"""
        if not self.__students:
            print("\n❌ No students in the system!")
            return

        print(f"\n{'Total Students in System: ' + str(len(self.__students)):^60}")
        for student in self.__students:
            student.display_details()

    def display_student_by_id(self, student_id):
        """Display details of a specific student"""
        student = self.find_student(student_id)
        if student:
            student.display_details()
        else:
            print(f"\n❌ Student with ID {student_id} not found!")

    def get_class_average(self):
        """Calculate average marks of all students"""
        if not self.__students:
            return 0

        total_avg = sum(student.get_average_marks() for student in self.__students)
        return round(total_avg / len(self.__students), 2)

    def get_topper(self):
        """Find the student with highest average marks"""
        if not self.__students:
            return None

        return max(self.__students, key=lambda s: s.get_average_marks())

    def display_statistics(self):
        """Display system-wide statistics"""
        if not self.__students:
            print("\n❌ No students in the system!")
            return

        print("\n" + "="*60)
        print(f"{'SYSTEM STATISTICS':^60}")
        print("="*60)
        print(f"Total Students       : {len(self.__students)}")
        print(f"Class Average Marks  : {self.get_class_average()}")

        topper = self.get_topper()
        if topper:
            print(f"Topper               : {topper.get_name()} ({topper.get_average_marks()})")
        print("="*60 + "\n")

# MAIN PROGRAM - DEMONSTRATION

def main():
    """Main function to demonstrate the Student Management System"""

    print("\n" + "="*60)
    print(f"{'STUDENT MANAGEMENT SYSTEM':^60}")
    print(f"{'OOP Concepts Demonstration':^60}")
    print("="*60)

    # Create management system
    sms = StudentManagementSystem()

    # --- Create and add students ---

    print("\n📚 Creating and adding students...")
    print("-" * 60)

    # Undergraduate student
    ug1 = UndergraduateStudent("UG001", "Alice Johnson", "Computer Science")
    ug1.set_marks("Math", 85)
    ug1.set_marks("Physics", 90)
    ug1.set_marks("Chemistry", 88)
    sms.add_student(ug1)

    # Postgraduate student
    pg1 = PostgraduateStudent("PG001", "Bob Smith", "Artificial Intelligence", "Neural Networks for NLP")
    pg1.set_marks("Advanced ML", 92)
    pg1.set_marks("Research Methods", 88)
    pg1.set_marks("Thesis Work", 95)
    sms.add_student(pg1)

    # International student
    int1 = InternationalStudent("INT001", "Emma Wilson", "Germany", "Valid")
    int1.set_marks("English", 78)
    int1.set_marks("Global Studies", 82)
    int1.set_marks("Intercultural Communication", 80)
    sms.add_student(int1)

    # Another undergraduate student
    ug2 = UndergraduateStudent("UG002", "Charlie Brown", "Business Administration")
    ug2.set_marks("Finance", 75)
    ug2.set_marks("Marketing", 80)
    ug2.set_marks("Economics", 72)
    sms.add_student(ug2)

    # Another postgraduate student
    pg2 = PostgraduateStudent("PG002", "Diana Prince", "Data Science", "Predictive Analytics in Finance")
    pg2.set_marks("Data Mining", 89)
    pg2.set_marks("Statistical Analysis", 91)
    pg2.set_marks("Big Data Technologies", 87)
    sms.add_student(pg2)

    # --- Display all students ---
    print("\n📋 Displaying all students:")
    sms.display_all_students()

    # --- Display specific student ---
    print("\n🔍 Searching for specific student (UG001):")
    sms.display_student_by_id("UG001")

    # --- Display statistics ---
    print("\n📊 System statistics:")
    sms.display_statistics()

    # --- Interactive menu ---
    print("\n" + "="*60)
    print("Interactive Menu (Optional)")
    print("="*60)

    while True:
        print("\nOptions:")
        print("1. Display all students")
        print("2. Search student by ID")
        print("3. Add marks to student")
        print("4. View statistics")
        print("5. Exit")

        choice = input("\nEnter your choice (1-5): ").strip()

        if choice == "1":
            sms.display_all_students()

        elif choice == "2":
            student_id = input("Enter Student ID: ").strip()
            sms.display_student_by_id(student_id)

        elif choice == "3":
            student_id = input("Enter Student ID: ").strip()
            student = sms.find_student(student_id)
            if student:
                subject = input("Enter Subject: ").strip()
                try:
                    marks = float(input("Enter Marks (0-100): ").strip())
                    student.set_marks(subject, marks)
                except ValueError:
                    print("❌ Invalid marks! Please enter a number.")
            else:
                print(f"❌ Student with ID {student_id} not found!")

        elif choice == "4":
            sms.display_statistics()

        elif choice == "5":
            print("\n👋 Thank you for using Student Management System!")
            break

        else:
            print("❌ Invalid choice! Please enter 1-5.")

if __name__ == "__main__":
    main()
```
