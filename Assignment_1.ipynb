```python
txt = "II'S ME PRADEEP"
# the lowercase converts string to lowercase
x = txt.lower()
print(x)
```

```python
txt = "it's pradeep"
# the uppercase converts string to uppercase
x = txt.upper()
print(x)
```

```python
txt = "i am a developer"
# the capitalize, capitalize the first letter
x = txt.capitalize()
print(x)
```

```python
txt = 'in life i need to be a "developer"'
# the title, capitalize the first letter of every word
x = txt.title()
print(x)
```

```python
txt = "   i dont't know how to be a '*********'.       "
# the strip Removes whitespace from the both ends
x = txt.strip()
print(x)
```

```python
txt = "     i am a man."
# the lstrip Removes whitespace from the left side
x = txt.lstrip()
print(x)
```

```python
txt = "i need to be a developer            "
# the rstrip Removes whitespace from the right side
x = txt.rstrip()
print(x)
```

```python
name = input("Enter your name: ")
# this is replace that replaes part os a string
# replace(old, new)
txt = f"my name is {name}"
x = txt.replace("pradeep", "over")
print(x)
```

```python
txt = "hello, developer"
# the split is use to splits string into a list using a separetor
x = txt.split()
print(x)
print(x[0])
```

```python
words = ["hello", "developer"]
# the join elements of a list with a string separetor
# join(iterable)
x = " ".join(words)
print(x)
```

```python
txt = "my name is pradeep"
# find is use for returns index of first occurrence, or -1 if not found
x = txt.find("is")
z = txt.find("over")
print(x)
print(z)
```

```python
txt = "my name is pradeep"
# index is use for returns index of first occurrence, but raises error if not found
x = txt.index("is")
print(x)
```

```python
txt = "i am a developer, i am a man "
# counts how many times a substring occurs
x = txt.count("i")
print(x)
```

```python
txt = "i am a live"
# checks if string starts with a substring
x = txt.startswith("i")
print(x)
```

```python
txt = "i am a live"
# checks if string ends with a substring
x = txt.endswith("live")
print(x)
```

```python
txt = "Hello123"
# Returns True if all characters are alphanumeric
# isalnum() just checks if a string contains only letters and numbers
# isalnum() does not allow gaps (spaces).
x = txt.isalnum()
print(x)
```

```python
txt = "Hello"
# Returns True if all characters are alphabetic
# isalpha() just checks if a string contains only letters
# isalpha() does not allow gaps (spaces).
x = txt.isalpha()
print(x)
```

```python
txt = "12334"
# Returns True if all characters are digits
# isdigit() just checks if a string contains only numbers
# isdigit() does not allow gaps (spaces).
x = txt.isdigit()
print(x)
```

```python
txt = "   "
# isspace() allow gaps (spaces). only
x = txt.isspace()
print(x)
```
